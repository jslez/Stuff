```python
## csvreader = csv.reader(uploaded['covid-19_tracking_survey_v7.csv'])
df = pd.read_csv(io.BytesIO(uploaded['covid-19_tracking_survey_v7.csv']))
```

```python
## See if the Tap Research-supplied gender data passed to us in the URL params matches the user's input to us directly

# Start by getting the relevant part of the dataframe
genders = []
genders = df[['questions.24','urlParams.2']].iloc[2:] ## pick these two labeled columns and chop off the first 2 rows
genders=genders.rename(columns={"questions.24": "swayable", "urlParams.2": "tap"})
```

```python
# Fix the convention for "M" / "male" and "F" / "female"
genders=genders.replace(to_replace="male",value="M")
genders=genders.replace(to_replace="female",value="F")
genders
```

```python
# Calculate the proportion of rows that have the same value in columns "swayable" and "tap"
sum(genders['swayable'] == genders['tap']) / len(genders)
```
